```python
#import important info (such as procedures to be used)
import math as maffs
import statistics as statz
```

```python
# Decision Tree Data Structure

# The Label Space is represented by a list of values representing the collection of every label 
class LabelSpace:
    def __init__(self, labels):
        self.labels = labels

### Labels for Assignment 1 are "+" and "-" ###
labelSpace1 = LabelSpace(["+", "-"])
### to keep track of the number of times each label occurs
labelNumOccur1 = LabelSpace([0, 0])
    
# A feature is represented by a list of types representing the collection of every value a feature can take
class Feature:
    def __init__(self, featureNumber, values):
        self.featureNumber = featureNumber
        self.values = values

### Features for Assignment 1 are of BOOLEAN type ###
# len(firstName) > len(lastName)
feature1 = Feature(1, [True,False])
# existsMiddleName?
feature2 = Feature(2, [True,False])
# firstName[0] == firstName[len(firstName) - 1]
feature3 = Feature(3, [True,False])
# min(listOfFirstAndLastName) == firstName?
feature4 = Feature(4, [True,False])
# firstName[1] in ["a","e","i","o","u","y"]
feature5 = Feature(5, [True,False])
# (len(lastName) % 2) == 0
feature6 = Feature(6, [True,False])
# APPLY line = line.lower() for case-insensitivity

# The Feature Space is represented by a list of types  representing the collection of every value a feature can take
class FeatureSpace:
    def __init__(self, features):
        self.features = features

### Feature Space for Assignment 1 is list containing lists with elements of BOOLEAN types ###
featureSpace = FeatureSpace([feature1, feature2, feature3, feature4, feature5, feature6])

# A Node for our tree can contain
# (String) label,
# (representation for Feature in FeatureSpace) featureIndex,
# (list of Nodes representing each decision) decisions
# node should be able to add possible Nodes to its list of possible decisions (for building tree)
class Node:
    def __init__(self, label, featureNumber, valueChosenForFeat = []):
        self.label = label
        self.featureNumber = featureNumber
        self.valueChosenForFeat = valueChosenForFeat
    def addNodeForFeatValue(self, chosenVal):
        self.valueChosenForFeat = self.valueChosenForFeat + [chosenVal]
        
# A Decision Tree starts from the root (i.e. a Node)
class DecisionTree:
    def __init__(self, iAmRoot):
        self.iAmRoot = iAmRoot
        
# An instance of data is represented by:
# (list of values for each Feature, in order added to Feature Space) featureValues = [], (String) label 
class Instance:
    def __init__(self, valuesOfFeatures, label):
        self.valuesOfFeatures = valuesOfFeatures
        self.label = label
        
# given a DecisionTree what label would we get for a given Instance 
def youGetALabel(cisionTree, Insta):
    # Start at root
    currNode = cisionTree.iAmRoot
    # while Node is not a leaf (i.e. no label)
    while ( len(currNode.label) == 0 ):
        # what decision are we taking (0 for True and 1 for False) found by answering:
        # what is the index for the current Node's feature (in the Feature Space)
        featIndex = currNode.featureNumber - 1
        # what value does that Feature have in our Instance
        instaFeatVal = Insta.valuesOfFeatures[featIndex]
        # what index does that value have in the Feature
        chosenValIndex = featureSpace.features[featIndex].values.index(instaFeatVal)
        # make decision decisionNum of our Node
        currNode = currNode.valueChosenForFeat[chosenValIndex]
    # out of while loop means currNode is a leaf (DecisionTree has put a label on the Instance)
    return currNode.label

# returns the proportion of correctly labeled Instances using a certain DecisionTree
def everyBodyGetsALabel(cisionTree, listOfInstas):
    numCorrectlyLabeled = 0
    for insta in listOfInstas:
        guessLabel = youGetALabel(cisionTree, insta)
        if ( insta.label == guessLabel ):
            numCorrectlyLabeled = numCorrectlyLabeled + 1
    return numCorrectlyLabeled/len(listOfInstas)
        
```

```python
# ID3 algorithm

# returns calculated entropy for a given list of Instances
def calcEntropyOf(listOfInstas):
    # set number of times we have seen each label to 0
    for label in labelNumOccur1.labels:
        label = 0
    # find out number of times each label occurs in the given list of Instances
    for Insta in listOfInstas:
        # from inside out
        # what is the label for the Instance
        # what is the index for that label in our Label Space
        # how many times have we seen that label
        labelNumOccur1.labels[labelSpace1.labels.index(Insta.label)] += 1
    
    sizeOfInstaList = len(listOfInstas)
    entropy = 0
    # calculate our entropy
    for index in range(len(labelSpace1.labels)):
        # find out the proportion of the label, at index index in the LabelSpace, in the listOfInstas
        labelPropo = labelNumOccur1.labels[index]/sizeOfInstaList
        entropy += (-1 * labelPropo * maffs.log2(labelPropo))
    return entropy
        
# returns calculated information gain of a certain Feature, in a list of Instances
def calcInformationGain(listOfInstas, featureNumb, entOfSamSpace):
    entOfFeat = 0
    
    # find entropy of each value that the (featureNum)th Feature can take on
    for featVal in featureSpace.features[featureNumb - 1].values:
        # find and store each Instance that has the value of featVal for the Feature: featureSpace[featureNum]
        instancesWithValOfInterest = []
        for insta in listOfInstas:
            if(insta.valuesOfFeatures[featureNumb - 1] == featVal):
                instancesWithValOfInterest.append(insta)
        # if no Instances with featVal were found, no calculation is needed
        if( len(instancesWithValOfInterest) <= 0 ):
            continue
        entropyOfFeatVal = calcEntropyOf(instancesWithValOfInterest)
        entOfFeat += ( (len(instancesWithValOfInterest) / len(listOfInstas)) * entropyOfFeatVal) 
        
    return entOfSamSpace - entOfFeat

# returns the index of the Feature with the Greatest value of Information Gain
def featWithGreatestInfoGain(listOfInstas, featSpace):
    maxFeatInfoGain = 0
    featIndx = 0
    
    entOfSampleSpace = calcEntropyOf(listOfInstas)
    
    # calculate Info. Gain of all Features and keep track of greatest value
    for feat in featSpace.features:
        iG = calcInformationGain(listOfInstas, feat.featureNumber, entOfSampleSpace)
        #print("iG of Feature #"+str(feat.featureNumber)+" is "+str(iG))
        if (iG > maxFeatInfoGain):
            maxFeatInfoGain = iG
            featIndx = feat.featureNumber
    return featIndx

# returns the label with the highest frequency in a given list of Instances
def mostSeenLabel(listOfInstas):
    # set number of times we have seen each label to 0
    for num in labelNumOccur1.labels:
        num = 0
    # find out number of times each label occurs in the given list of Instances
    for Insta in listOfInstas:
        # from inside out
        # what is the label for the Instance
        # what is the index for that label in our Label Space
        # how many times have we seen that label
        labelNumOccur1.labels[labelSpace1.labels.index(Insta.label)] += 1
        
    return labelSpace1.labels[labelNumOccur1.labels.index(max(labelNumOccur1.labels))]

#ID3 algorithm forms DecisionTree from a given list of Instances, a FeatureSpace, and a target label
# by returning one Node at a time
def ID3(listOfInstas, featSpace, commLabel, depthLimit = -1, depthMeasure = 0):
     
    # check if all Instances have the same label
    if (len(listOfInstas) == 1):
        #print("len(listOfInstas) == 1")
        return Node( commLabel, -1 )
    isHomogenous = True
    for indx in range(len(listOfInstas)-1):
        if (not ( listOfInstas[indx].label == listOfInstas[indx+1].label ) ):
            isHomogenous = False
            break
    # if so return single (leaf) Node with that label
    if isHomogenous:
        #print("isHomogenous")
        return Node( commLabel, -1 )
    
    # we have worked through all Features or depthMeasure has reached a positive depth
    if ( (len(featSpace.features) < 1) or (depthLimit > 0 and depthMeasure == depthLimit)):
        return Node( commLabel, -1 )
    
    ## different labels means we need to create a node with the Feature that best classifies the listOfInstas
    ### in this case the best Feature is the one with the most information gain
    # index in FeatureSpace of best Feature classifier 
    bestClassifierIndex = featWithGreatestInfoGain(listOfInstas, featSpace)
    #print("bestClassifierIndex = "+str(bestClassifierIndex))
    nodeToBeAddedToTree = Node( "", bestClassifierIndex )
    
    properIndex = 1
    ### find the index of the best Classifier in our featSpace
    for feat in featSpace.features:
        if (feat.featureNumber == bestClassifierIndex):
            break
        else:
            properIndex += 1
    #print("properIndex = "+str(properIndex))
        
    # for each value that the best Feature can take
    for val in featSpace.features[properIndex-1].values:
        # find all Instances whose Feature, the one we consider as best classifier, takes on the value of val
        instasWithValOfInterest = []
        for insta in listOfInstas:
            if (insta.valuesOfFeatures[properIndex-1] == val):
                instasWithValOfInterest.append(insta)
                
        # if our list of Instances contains no Instance where the Feature takes on the value of val
        # then we add a leaf node with the most common label seen in the list (for generalization)
        if( len(instasWithValOfInterest) == 0 ):
            nodeToBeAddedToTree.addNodeForFeatValue( Node( commLabel, -1) )
            
        # else add subtree for instasWithValOfInterest and (Featurespace - featSpace[bestClassifierIndex-1])
        else:
            commonLabel = mostSeenLabel(instasWithValOfInterest)
            featSpaceFeatsCopyMinusOne = list(featSpace.features)
            del featSpaceFeatsCopyMinusOne[properIndex-1]
            featSpaceCopyMinusOne = FeatureSpace(featSpaceFeatsCopyMinusOne)
            nodeToBeAddedToTree.addNodeForFeatValue( ID3(instasWithValOfInterest, featSpaceCopyMinusOne, commonLabel, depthLimit, depthMeasure) )
    
    # node is ready to go
    return nodeToBeAddedToTree
```

```python
# parse data from given files 
### Data files for Assignment 1 are of the form: ###
### (label) (name) ###

# goal is to append Instances from a given file into a list of Instances and return the resulting list of Instances
def whatAmILookingAt(file, listOfInstas):
    phile = open(file, "r")
    lines = phile.readlines()
    
    for line in lines:
        lowLine = line.lower()
        
        labelFound = lowLine[0]
        
        wholeName = lowLine[2:]
        nameTerms = wholeName.split()
        
        firstName = nameTerms[0]
        lastName = nameTerms[-1]
        middleName = ""
        # is there more than the first name term and last name term
        if ( len(nameTerms) > 2 ):
            middleName = " ".join(nameTerms[1:-1])
            # since there could be multiple middle terms in the name (as seen in the data ex. "van der")
        
        # Now we define the value for each of our features for this particular Instance (line in lines)
        valOfFeat1 = len(firstName) > len(lastName)
        valOfFeat2 = len(middleName) > 0
        valOfFeat3 = firstName[0] == firstName[-1]
        valOfFeat4 = min([] + [nameTerms[0]] + [nameTerms[-1]]) == firstName
        if ( len(firstName) > 1 and firstName[1].isalpha() ):
            valOfFeat5 = firstName[1] in ["a","e","i","o","u"]
        else:
            valOfFeat5 = False
        valOfFeat6 = (len(lastName) % 2) == 0
        
        # create Instance with found values of each Feature
        instaFound = Instance([valOfFeat1, valOfFeat2, valOfFeat3, valOfFeat4, valOfFeat5, valOfFeat6], labelFound)
        # append Instance we found to our list of Instances
        listOfInstas.append(instaFound)
        
    return listOfInstas

# parse training data file
trainingFile = "DATA/training.txt"
trainingListOfInstas = whatAmILookingAt(trainingFile, [])
# parse testing data file
testingFile = "DATA/test.txt"
testingListOfInstas = whatAmILookingAt(testingFile, [])
    
```

```python
# parse the given 4 cross-validation files 

# parse first fold data file
foldTrainingFile00 = "DATA/folds/training-00.txt"
fold00ListOfInstas = whatAmILookingAt(foldTrainingFile00, [])
# parse second fold data file
foldTrainingFile01 = "DATA/folds/training-01.txt"
fold01ListOfInstas = whatAmILookingAt(foldTrainingFile01, [])
# parse third fold data file
foldTrainingFile02 = "DATA/folds/training-02.txt"
fold02ListOfInstas = whatAmILookingAt(foldTrainingFile02, [])
# parse fourth fold data file
foldTrainingFile03 = "DATA/folds/training-03.txt"
fold03ListOfInstas = whatAmILookingAt(foldTrainingFile03, [])
folds = [fold00ListOfInstas,fold01ListOfInstas,fold02ListOfInstas,fold03ListOfInstas]
```

#### (a) 

```python
# Running 4-fold cross-validation using the specified files
# k = number of folds
# Experiment with depths in the set {1,2,3,4,5,10,15,20} (choices of hyperparameter)
hyperparameterChoices = [1,2,3,4,5,10,15,20]

# track of which depth has highest average accuracy
bestChoice = 0
highestAvgAcc = 0

for choice in hyperparameterChoices:
    accuracyForTestFold = []
    ## Train k times (to evaluate each node) for each depth 
    for testIndex in range(len(folds)):
        foldsToTrainOn = []
    ### On k - 1 folds 
        for trainIndex in range(len(folds)):
            if (not (testIndex == trainIndex)):
                foldsToTrainOn.extend(folds[trainIndex])
        # train DecisionTree on foldsToTrainOn within featureSpace
        # passing the mostSeenLabel in foldsToTrainOn and a depthLimit of depth
        decTreeFold = DecisionTree(ID3(foldsToTrainOn, featureSpace, mostSeenLabel(foldsToTrainOn), choice))
        # calc accuracy of Tree and keep track
        accuracyForTestFold.append(everyBodyGetsALabel(decTreeFold, foldsToTrainOn))
        
    sumOfAccuracies = sum(accuracyForTestFold)
    avgAcc = sumOfAccuracies/len(folds)
    standDev = statz.stdev(accuracyForTestFold)
    if (avgAcc > highestAvgAcc):
        bestChoice = choice
        highestAvgAcc = avgAcc
    # Reporting the cross-validation accuracy and standard deviation for each depth.
    print("========================")
    print("For a depth limit of "+str(choice))
    print("Average Accuracy is "+str(avgAcc))
    print("Standard Deviation is "+str(standDev))
    
# Explicity specifying which depth should be chosen as the best, and explaining why.
print("The best choice for our Hyperparameter is the value of "+str(bestChoice)+".")
print("This is the best choice because it has the highest average accuracy of "+str(highestAvgAcc)+" out of all of the choices.")
```
